## Investigation Overview

In this Investigation , I will be focusing on salaries of the data science jobs, The highest paying jobs , the average in the data science industry , the ratio of remote working , how the experience


## Dataset Overview

This is a kaggle dataset downloaded via API and Extracted. The dataset has 607 rows and 11 columns. The columns range from year to salary to job loaction and other feature. The dataset has no nulls and it is tidy in terms of structure

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('ds_salaries.csv',index_col=False)
```

```python
#Remove unnamed column
df = df.iloc[: , 1:]
```

```python
#Convert to necessary dTypes for analysis
df.work_year.astype('int32')
df['work_year'] = pd.to_datetime(df.work_year, format = '%Y')
df['salary']  =  df['salary'].astype('float')
df['salary_in_usd']  =  df['salary_in_usd'].astype('float')
```

```python
#Change multiple columns to correct dtye
for col in ['experience_level', 'employment_type', 'company_size',]:
    df[col] = df[col].astype('category')
```

```python
#Change multiple columns to correct dtye
for col in ['job_title', 'company_location', 'employee_residence','salary_currency']:
    df[col] = df[col].astype('string')
```

## Average salary(in usd) in data science industry over the years 
This graph shows the average for the data science jobs and it clearly shows that the average salary has been increasing each year and there was rather a drastic increase from 2021 to 2022 with an increase of 20K




```python
df.groupby(df['work_year'].dt.year)['salary_in_usd'].mean().plot(kind='bar');
plt.title("Average salary is USD over the years",fontsize=25);
plt.xlabel("Year",labelpad=5,fontsize=20)
plt.ylabel("Salary in USD($)",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.yticks([20000,40000,60000,80000,100000,120000],['20k','40k','60k','80k','100k','120k']);
plt.gcf().set_size_inches(14, 9)
```

## Country with the most job entries
This graph clearly shows that USA has the most job entries with above 50% of the data coming from the US while a Great britain with second contributing about 8% of the data and the rest of the countries below 8%

```python
### Create Function for poltiing bar charts
def plotbarchat(dfcolumn,dataframe,column,colorx,title): 
                   base_color = sb.color_palette()[colorx]
                   freq = dfcolumn.value_counts().index
                   sb.countplot(data=dataframe, x=f"{column}", color=base_color, order=freq)
                   plt.title(f"{title}")
```

```python
plotbarchat(df['employee_residence'],df,"employee_residence",1,"Top ten Countries with most job entries")
plt.xticks(rotation=90)
plt.xlim((0,10));
```

## The job which pays the most in the Data Science industry between 2020 and 2022 in usd
This graph depicts the highest or maximum salary for each job between the year 2020 to 2022. The top 3 highest earners are 
<ol>
    <li>Principal Data Engineer with over 600k per annum</li>
    <li>Research Scientist and Financial Data Analyst above 400k per annum</li>
    <li>Applied Machine Learning Scientist also above the 400k per annum </li>
</ol>



```python
df.groupby(['job_title'])['salary_in_usd'].max().plot(kind='bar')
plt.title("Maximum salary in USD($) for each job title",fontsize=20)
plt.margins(x=0, y=0)
plt.xlabel("Job Titles", labelpad=7,fontsize=20)
plt.ylabel("Salary in USD($)",fontsize=15)
plt.yticks([100000,200000,300000,400000,500000,600000],['100k','200k','300k','400k','500k','600k'],fontsize=25)
plt.xticks(fontsize=25)
plt.figure(figsize=[14, 8])
plt.show()
```

## The remote ratio each job
This focuses on each job title and the number of remote ratios jobs for each job title. In this graph the job title with the highest remote position entries is Data engineer. But seeing the graph its easy to tell that all the job titles have remote position as their highest. Meaning most data science position are remote located



```python
g= sb.countplot(data=df,x='job_title',hue='remote_ratio')
plt.title("Remote status of jobs",fontsize=25)
plt.ylabel("Job Count",fontsize=20)
plt.xlabel("Job Title",fontsize=20)
plt.xticks(fontsize=20)
plt.xticks(rotation=90)
plt.yticks(fontsize=20)
sb.move_legend(g, "upper center")
plt.gcf().set_size_inches(14, 9)
plt.show()
```

## The salaries based on remote ratio in different company sizes
This graphic shows the three company size and remote ratio for each company size as well as how much each remote ratio position earns in each company. In large sized companies , Remote data science Jobs get the highest pay and that is true as well for small and medium sized companies

```python
ax = sb.barplot(data = df, x = 'company_size', y = 'salary_in_usd',
                hue = 'remote_ratio')
ax.legend(loc='upper center', ncol = 3, framealpha = 1, title = 'Remote Ratio')
ax.set_title("Remote Ratio salaries at different company size",fontsize=30)
plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='20')
plt.yticks([20000,40000,60000,80000,100000,120000,140000,160000],
           ['20k','40k','60k','80k','100k','120k','140k','160k'],fontsize=15)
plt.xticks(fontsize=15)
plt.xlabel("Company size", fontsize=20)
plt.ylabel("Salary in USD",fontsize=20)
plt.gcf().set_size_inches(14, 9);
```

## Measuring salary by experience level over time
The experience level is key feature which determines how much you would earn. The highest earning experience level is Executive level followed by the senior level and then the middle (intermediate) and entry position. There is an decrease in 2021 for Senior ,middle and entry position while executive had its all time high in 2021. And whilst executive salaries decrease in 2022 , the other exprience levels have a slight increase going into 2022

```python
ax = sb.lineplot(x="work_year", y="salary_in_usd",hue="experience_level" ,
    markers=True,dashes=False,ci=0,data=df)
ax.set_title("Salary Vs Experience level over time",fontsize=30)
plt.yticks([75000,100000,125000,150000,175000,200000,225000],
          ['75K','100K','125K','150K','175K','200K','225K'],fontsize=20)
plt.xticks(fontsize=20)
plt.xlabel("Year", fontsize=20)
plt.ylabel("Salary in USD",fontsize=20)
plt.gcf().set_size_inches(14, 9);
```
