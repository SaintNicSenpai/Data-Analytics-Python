#### Question 1
What is the frequency distribution in the work experience column?Its a qualitative variable(categorical)  so I prefer to use a bar chat

#### Code and Visualization

```python
base_color = sb.color_palette()[1]

freq = df['experience_level'].value_counts().index


# Plot the bar chart in the decreasing order of the frequency experience level`
sb.countplot(data=df, x='experience_level', color=base_color, order=freq)
plt.title("Experience Level Frequency in dataset")
```

#### Insights
Most jobs are of Senior level experienece and the least is being te executive/Expert level. Meaning most of the jobs were not for expert/executive positions

```python
### Create Function for poltiing bar charts
def plotbarchat(dfcolumn,dataframe,column,colorx,title): 
                   base_color = sb.color_palette()[colorx]
                   freq = dfcolumn.value_counts().index
                   sb.countplot(data=dataframe, x=f"{column}", color=base_color, order=freq)
                   plt.title(f"{title}")
```

#### Question 4

```python
#Since we are working with millions and thousands the log binsize is needed
log_binsize = 0.5
bins = 10 ** np.arange(2.4, np.log10(df['salary'].max())+log_binsize, log_binsize)

plt.figure(figsize=[12, 11])
plt.hist(data = df, x = 'salary', bins = bins)
plt.xscale('log')
plt.xticks([4e3, 3e6, 6e6, 9e6, 1e7,1.e7, 1.e7, 2e7, 2e7, 2e7, 3e7], ['4k', '3M', '6M', '9M', '10M', '10M','10M','20M', '20M', '20M', '30M'],rotation =90)
plt.xlabel('Salaries($)')
plt.show()
```

####  Insights
According to the data almost of the salaries with their different currencies fall between 4k and 3M.The data skews to the right and we have a few outlier millions 20m or 30m recorded

## Bivariate Exploration
In this part we'l be exploring casual and relationship between our main variable and the other variables variables

#### Question 1 Has the average salary increased or decreased ?

#### Code and Visualization

```python
df.groupby(df['work_year'].dt.year)['salary_in_usd'].mean().plot(kind='bar');
plt.title("Average salary is USD over the years");
plt.xlabel("Work year", labelpad=7)
plt.ylabel("Salary in USD($)");
```

#### Insight
The average salary increased has increased from 2020 to 2022 a gradual increase

#### Question 2
What is the maximum salary for each Job title in USD?

#### Code and Visualization

```python
df.groupby(['job_title'])['salary_in_usd'].max().plot(kind='bar')
plt.title("Maximum salary in USD($) for each job title")
plt.rcParams["figure.figsize"] = (15,10)
plt.rcParams["figure.autolayout"] = True
plt.margins(x=0, y=0)
plt.xlabel("Job Titles", labelpad=7,fontsize=20)
plt.ylabel("Salary in USD($)",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
log_binsize = 0.5
df.groupby(['job_title'])['salary'].max().plot(kind='bar')
plt.yscale('log')
plt.title("Maximum salary for each job title in Different currencies")
plt.rcParams["figure.figsize"] = (15,10)
plt.rcParams["figure.autolayout"] = True
plt.margins(x=0, y=0)
plt.xlabel("Job Titles", labelpad=7,fontsize=20)
plt.ylabel("Salary in Different currencies",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

#### Insights
There are two bar graphs which show the maximum salary for each job title. The first graph shows the maximum in usd and the other in all the other currencies. There's clearly different maximum salaries for the same job titles when comparing the two graphs. Also the y scale difference , the other has it maximum at 10 to the power 7 which is 10 million whilst the other has an upper limit of 6 000 000

#### Question 4

#### Question
Which company size has the most remote jobs?

#### Code and Visualize

```python
ax = sb.barplot(data = df, x = 'company_size', y = 'salary_in_usd',
                hue = 'remote_ratio')
ax.legend(loc='upper center', ncol = 3, framealpha = 1, title = 'Remote Ratio')
ax.set_title("Remote Ratio salaries at different company size",fontsize=30)
plt.setp(ax.get_legend().get_texts(), fontsize='15') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='20')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Company size", fontsize=20)
plt.ylabel("Salary in USD",fontsize=15);
```

####  Insight
The Large companies have the highest paid remote workes , followed by medium followed by small companies. For 50% Remote  The large companies have the highest followed by the small companies
